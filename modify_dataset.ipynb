```python
import os
import shutil
```

```python
# Rename file_name from before to after
# ex. renamefile(file_name, "JPG", "jpg")

def rename_file(file_name, before, after):
    new_name = file_name.replace(before, after)
    shutil.move(file_name,new_name)
    print("Renamed", file_name, " to", new_name)
```

```python
# Delete file
# ex. delete_file("C:\User\test.txt")

def delete_file(file_path):
    os.remove(file_path)
```

```python
# Move file
# ex. move_file("C:\User\text.txt", "C:\User\other_directory\test.txt")

def move_file(src,des):
    shutil.move(src, des)
```

```python
name_to_id = {
    "plastic_bottle":0,
    "bottle":0,
    "fork":1,
    "knife":2,
    "spoon":3
}
```

```python
# convert names to ids in txt files

def names_to_ids(file_name):
    file = open(file_name,'r')
    output = ""
    for line in file:
        line = line.split()
        line[0] = str(name_to_id[line[0]])
        output += create_str(line)
    file.close()
    
    print("Replacing name with id")
    
    file = open(file_name,'w')
    file.write(output)
    file.close()
```

```python
# Modify txt file

def modify_txt(file_name, object_id):
    file = open(file_name, 'r')
    output = ""
    for line in file:
        line = line.split()
        line[0] = str(object_id)
        output += create_str(line)
    file.close()

    print("Writing to file")
    
    file = open(file_name,'w')
    file.write(output)
    file.close()
    
def create_str(L):
    output = ""
    for i in L:
        output += i
        output += ' '
    output += '\n'
    return output
```

```python
def main():
    directory = "data_MASTER"
    for i in os.listdir(directory):
        '''
        if i.endswith("coords.txt"):
            delete_file(os.path.join(directory,i))
        '''
        
        if i.endswith(".txt"):
            names_to_ids(os.path.join(directory,i))
        else: # If jpg does not have txt, delete jpg
            if os.path.exists(os.path.join(directory,i.replace("jpg","txt"))) == False:
                delete_file(os.path.join(directory,i))
        
```

```python
main()
```
